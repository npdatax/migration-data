### Profession Tree

```python
import json
```

```python
with open("profession.json", 'r') as f5:
    profession = json.load(f5)
```

```python
i = 1;j=1;k=1

Data = []
Data.append({"id" : "profession","value":str(i)})

for key,value in profession.items():
        k = 1
        if len(key)<10000:
            row = "profession" + "." +  key
            Data.append({"id": row,"value":str(i) + "." + str(j)})
        
            for t in value:
                if len(t)>1 and len(t) < 10000:
                    if t[-1] == "\r":
                        t = t[0:-1]
                    nrow = row + "." +t
                    Data.append({"id":nrow,"value":str(i) + "." + str(j) + "." +  str(k)})
                    k = k+1
        j = j+1
```

```python
Data
```

```python
import pandas as pd
DF = pd.DataFrame(Data)
DF = DF.set_index("id")
```
