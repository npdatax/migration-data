### Profession Generator

```python
import json as json
import random as random
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
sns.set()
```

```python
pf = pd.read_csv("profession.csv", header = None )
pf.shape
```

```python
Proff = {}
for row in range(pf.shape[0]):
    major = str(pf.loc[row,0])
    if major[0] ==" ":
        major = major[1:]
    branches = str(pf.loc[row,1])
    brn = branches.split("\n")
    BRN = []
    for item in brn:
        if len(item) >2:
            if item[0] == " ":
                #print("yes")
                item = item[2:]
            if item[-1] == "\r":
                item = item[0:-1]
        BRN.append(item)   
    Proff.update({major:BRN})
    for item in BRN:
        print(major,"|",item)
```

```python
with open("profession.json",'w') as f:
    json.dump(Proff,f)
```

### Generate random Profession

```python
all_professions = list(Proff.keys())
for k in range(20):
    major = random.choice(all_professions)
    branch = random.choice(Proff[major])
    print(major,"|",branch)
```
