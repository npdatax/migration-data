### Geographic Data Generator

```python
import json as json
import numpy as np
import random as random
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
sns.set()
```

#### Countries

```python
with open("world-countries-folium.json", 'r') as fw:
    wc = json.load(fw)
```

#### Cities

```python
countries = []
for item in wc['features']:
    c = item['properties']['name']
    ID = item['id']
    countries.append({"name" : c, "ID":ID})
```

```python
with open("World_countries.json", 'w') as f:
    json.dump(countries,f)
```

```python
CNT = pd.DataFrame(countries)
CNT = CNT.set_index('ID')
CNT.head()
```

```python
CNT.loc['AFG','name']
```

#### Generate 1000 data

```python
def generate_foreign_address(countries):
    c = random.choice(countries)
    return c
```

```python
generate_foreign_address(countries)
```

```python
Cdict = []
for k in range(100000):
    country  = generate_foreign_address(countries)
    Cdict.append(country)
```

```python
DF = pd.DataFrame(Cdict)
DF = DF.set_index('ID')
DF.head()
```

```python
df1 = DF.groupby('ID').count()
df1.head()
```

```python
CNT['count'] = np.zeros(CNT.shape[0])
for item in list(CNT.index):
    CNT.loc[item,'count'] = df1.loc[item,'name']%200
```

#### Geographic Visualization (Google map API)

```python
import folium as folium
```

```python
# Make an empty map
m = folium.Map([25,-10], zoom_start=2)

state_geo = 'data/world-countries.json'


# add json data
m.choropleth(geo_data=state_geo,\
            data=CNT,\
            columns=['name', 'count'],\
            key_on='feature.properties.name',\
            fill_color='YlGn',\
            fill_opacity=0.7,\
            line_opacity=0.2,\
            legend_name='rate of publication')

# Save it as html
m.save('world.html')
#m.save('plot.pdf')
```

```python
import folium

m = folium.Map([51.5, -0.25], zoom_start=10)

test = folium.Html('<b>Hello world</b>', script=True)

popup = folium.Popup(test, max_width=2650)
folium.RegularPolygonMarker(
    location=[51.5, -0.25], popup=popup,
).add_to(m)

m.save('osm.html')
```
